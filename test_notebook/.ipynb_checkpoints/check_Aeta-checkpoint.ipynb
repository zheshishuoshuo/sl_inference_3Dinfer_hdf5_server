```python
import h5py
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# === 参数设置 ===
bank_file = "../bank_uniform.h5"      # 你的 bank 文件
muDM_grid = np.linspace(12.0, 14.5, 26)  # 按需要修改 μ_DM 取值
betaDM = 2.0
sigmaDM = 0.37
alpha = 0.15

# === 读取数据 ===
with h5py.File(bank_file, "r") as f:
    logMh = f["/base/logMh"][:]
    logMstar = f["/base/logM_star"][:]
    logRe = f["/base/logRe"][:]
    betamax = f["/lens/betamax"][:]
    mu1 = f["/lens/mu1"][:]
    mu2 = f["/lens/mu2"][:]
    logq_Mh = f["/weights/logq_Mh"][:]
    logq_Ms = f["/weights/logq_Mstar"][:]
    logq_Re = f["/weights/logq_Re"][:]

# === 构造静态权重（不依赖 μ_DM） ===
q_Mh = np.exp(logq_Mh)
q_Ms = np.exp(logq_Ms)
q_Re = np.exp(logq_Re)

# 这里可以用更精确的 p(Msps) 和 p(Re|Msps)，为了诊断先只看 q 的影响
w_static = 1.0 / (q_Mh * q_Ms * q_Re)

# === 循环每个 μ_DM 计算 p(logMh|μ_DM) 和有效样本数 ===
N_eff_list = []
max_w_list = []
mean_w_list = []

for muDM in muDM_grid:
    Msps = logMstar - alpha
    loc = muDM + betaDM * (Msps - 11.4)
    p_mh = norm.pdf(logMh, loc=loc, scale=sigmaDM)

    w_eff = w_static * p_mh

    w_sum = np.sum(w_eff)
    w_sq_sum = np.sum(w_eff**2)

    if w_sq_sum > 0:
        N_eff = (w_sum**2) / w_sq_sum
    else:
        N_eff = 0.0

    N_eff_list.append(N_eff)
    max_w_list.append(np.max(w_eff))
    mean_w_list.append(np.mean(w_eff))

# === 打印统计结果 ===
print("μ_DM   N_eff     max/mean ratio")
for muDM, Neff, maxw, meanw in zip(muDM_grid, N_eff_list, max_w_list, mean_w_list):
    ratio = maxw / (meanw + 1e-30)
    print(f"{muDM:5.2f}  {Neff:8.1f}  {ratio:10.1f}")

# === 可视化 ===
plt.figure(figsize=(6,4))
plt.plot(muDM_grid, N_eff_list, marker="o", label="N_eff")
plt.xlabel("μ_DM")
plt.ylabel("Effective sample size")
plt.title("Effective sample size vs μ_DM")
plt.grid()
plt.legend()
plt.show()
```

```python
import h5py
import numpy as np
from scipy.interpolate import RegularGridInterpolator

def load_p_mstar_interp(path):
    with h5py.File(path, "r") as f:
        keys = list(f.keys())
        # 应该能看到 ['alpha_grid', 'mstar_grid', 'p_mstar']
        # 老版本可能是 'p_mstar_table' 或 'p_table'
        # 逐个尝试，避免 KeyError
        if "p_mstar" in keys:
            pkey = "p_mstar"
        elif "p_mstar_table" in keys:
            pkey = "p_mstar_table"
        elif "p_table" in keys:
            pkey = "p_table"
        else:
            raise KeyError(f"No p(M★|α) dataset found; available keys: {keys}")

        alpha_grid = np.array(f["alpha_grid"], dtype=np.float64)
        mstar_grid = np.array(f["mstar_grid"], dtype=np.float64)
        p_table    = np.array(f[pkey],        dtype=np.float64)  # shape (n_alpha, n_mstar)

    # 构造二维插值器：(alpha, mstar) -> p(M★|α)
    return RegularGridInterpolator(
        (alpha_grid, mstar_grid), p_table,
        bounds_error=False, fill_value=0.0
    )

# from utils import load_p_mstar_interp  # 或直接把刚才的函数粘贴到你的脚本

p_mstar_interp = load_p_mstar_interp("../p_mstar_table.h5")
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# === 读取新表 (4D)，只取 α=0.15 ===
def load_new_aeta_1d(filename, alpha_target=0.15):
    with h5py.File(filename, "r") as f:
        muDM = f["mu_dm_grid"][:]
        alpha_grid = f["alpha_grid"][:] if "alpha_grid" in f else np.array([alpha_target])
        A_eta = np.squeeze(f["A_eta"][:], axis=(1, 2))  # shape (n_muDM, n_alpha)
    ia = np.argmin(np.abs(alpha_grid - alpha_target))
    return muDM, A_eta[:, ia]

# === 读取旧表 (grids 组) ===
def load_old_aeta_1d(filename):
    with h5py.File(filename, "r") as f:
        muDM = f["grids/mu_DM_grid"][:]
        A_eta = f["grids/A_grid"][:]
    return muDM, A_eta

# === 加载 & 对齐 ===
mu_new, A_new = load_new_aeta_1d("../../Aeta_uniform_grid_with_pmstar.h5", alpha_target=0.15)
mu_to, A_to = mu_new, A_new  # 直接用新表
mu_old, A_old = load_old_aeta_1d("../aeta_tables/Aeta_20250906T041524Z.h5")
A_old_interp = np.interp(mu_new, mu_old, A_old)

# # === 数量级对齐：在参考区间归一化 ===
ref_mask = (mu_new >= 12) & (mu_new <= 14)
scale = np.mean(A_old_interp[ref_mask]) / np.mean(A_new[ref_mask])
A_new_scaled = A_new * scale

# 计算相对偏差
rel_diff = (A_new_scaled - A_old_interp) / A_old_interp

print(f"归一化因子 scale: {scale:.3e}")
print(f"平均相对偏差: {np.nanmean(rel_diff):.3e}")
print(f"最大相对偏差: {np.nanmax(np.abs(rel_diff)):.3e}")

# === 可视化对比 ===
plt.figure(figsize=(7, 5))
# plt.plot(mu_new, A_new_scaled, label="New A(η) α≈0.15 (scaled)")
plt.scatter(mu_new, A_new_scaled, label="New A(η) α≈0.15 (scaled)", s=10)
plt.plot(mu_new, A_old_interp, "--", label="Old A(η) interp")
plt.xlabel(r"$\mu_{\rm DM}$")
plt.ylabel(r"$A(\eta)$")
plt.title("New vs Old A(η) (α=0.15, scaled)")
plt.legend()
plt.grid()
plt.show()

# === 可视化相对差异 ===
plt.figure(figsize=(7, 3))
plt.plot(mu_new, rel_diff * 100)
plt.axhline(0, color="k", ls="--")
plt.ylabel("Relative Diff [%]")
plt.xlabel(r"$\mu_{\rm DM}$")
plt.title("Relative Difference After Scaling (New - Old)/Old")
plt.grid()
plt.show()
```

```python
import h5py
import numpy as np
from tqdm import tqdm

# === 参数设置 ===
grid_file = "../grid_mock_significant.h5"
N_sample = 1_000_000  # 想要抽取的样本数量
chunk_size = 100_000  # 每次读取多少行
rng = np.random.default_rng(seed=42)

with h5py.File(grid_file, "r") as f:
    N_total = len(f["base/logMh"])
    keep_prob = min(1.0, N_sample / N_total)

    logMh_all, logMstar_all, logRe_all = [], [], []
    beta_all, mu1_all, mu2_all, bmax_all = [], [], [], []

    for start in tqdm(range(0, N_total, chunk_size), desc="Scanning grid"):
        end = min(start + chunk_size, N_total)
        # 顺序读一块
        logMh_chunk = f["base/logMh"][start:end]
        mask = rng.random(len(logMh_chunk)) < keep_prob
        if np.any(mask):
            logMh_all.append(logMh_chunk[mask])
            logMstar_all.append(f["base/logMstar"][start:end][mask])
            logRe_all.append(f["base/logRe"][start:end][mask])
            beta_all.append(f["base/beta"][start:end][mask])
            mu1_all.append(f["lens/mu1"][start:end][mask])
            mu2_all.append(f["lens/mu2"][start:end][mask])
            bmax_all.append(f["lens/betamax"][start:end][mask])

# 拼接成一个样本子集
logMh = np.concatenate(logMh_all)
logMstar = np.concatenate(logMstar_all)
logRe = np.concatenate(logRe_all)
beta_unit = np.concatenate(beta_all)
mu1 = np.concatenate(mu1_all)
mu2 = np.concatenate(mu2_all)
betamax = np.concatenate(bmax_all)

print(f"[info] 抽取了 {len(logMh)} / {N_total} 个样本 ({100*len(logMh)/N_total:.3f}%)")

# === 打印物理量统计 ===
def summarize(name, arr):
    arr = arr[np.isfinite(arr)]
    print(f"{name:<8}: mean={np.mean(arr):.3f}, std={np.std(arr):.3f}, min={np.min(arr):.3f}, max={np.max(arr):.3f}")

summarize("logMh", logMh)
summarize("logMstar", logMstar)
summarize("logRe", logRe)
summarize("beta", beta_unit)
summarize("mu1", mu1)
summarize("mu2", mu2)
summarize("betamax", betamax)
```
